```python
#query data for the last X number of days from today. This is used to calculate Exponential Mean Average (EMA)
queryStartDate = (dt.datetime.today() - dt.timedelta(days=50)).date()
queryEndDate = (dt.datetime.today() - dt.timedelta(days=10)).date()
query_daily_price_sql = '''SELECT ticker, close_price from INDEX_DAILY_PRICE 
    where close_date > "''' + str(queryStartDate) + '''" and close_date < "''' + str(queryEndDate) + '''"'''
c.execute(query_daily_price_sql)
result = c.fetchall()
print(result[0:3])
```

```python
def conversion(tup, dict):
    for x, y in tup:
        dict.setdefault(x, []).append(y)
    return dict
```
